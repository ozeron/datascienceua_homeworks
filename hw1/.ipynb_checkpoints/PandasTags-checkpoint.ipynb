```python
def clean_values(string):
    for ch in "[], '":
        string = string.replace(ch, ' ')
    return string.split()
```

```python
df = pandas.read_csv('test.csv')
df.columns = ['date_published', 'author', 'comments_count', 'tags', 'title']
df['tags'] = map(clean_values, df['tags'])
```
